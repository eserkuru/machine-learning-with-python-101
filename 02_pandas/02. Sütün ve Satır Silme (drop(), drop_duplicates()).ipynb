# drop() ile sütun silmek

```python
df.drop('yas', axis=1, inplace=True)
# inplace True işlemlerini bir defa yapabiliriz. Çünkü bellekte veri değişiyor.
# Yukarıda kulandığımız axis=1 silmenin sütun bazlı olacağını inplace=True ise 
# silmenin gerçekten bu dataframe üzerinde gerçekleşeceğini gösterir
```

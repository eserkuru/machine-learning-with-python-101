matplotlib, birçok yeni kullanıcı için göz korkutucu olabilecek kapsamlı bir koda sahiptir. Bununla birlikte, matplotlib'in çoğu, oldukça basit bir kavramsal çerçeve ve birkaç önemli nokta bilgisi ile anlaşılabilir.

Bu paketin amacı, verilerinizi olabildiğince kolay bir şekilde görselleştirmenize yardımcı olmaktır.

matplotlib'deki her şey bir hiyerarşide düzenlenir. Hiyerarşinin en üstünde, matplotlib.pyplot modülü tarafından sağlanan matplotlib "state-machine environment" dır. 

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

%matplotlib inline
```

```python
plt.plot()



plt.show()
```

# Eksen İsimlerini Eklemek

```python
plt.plot()

plt.xlabel("Burası x ekseni")
plt.ylabel("Burası y ekseni")

plt.show()
```

# Başlık Eklemek

```python
plt.plot()
plt.xlabel("Burası x ekseni")
plt.ylabel("Burası y ekseni")

plt.title("Burası başlık")
plt.show()
```

# Legend Ekleme

```python
plt.plot()
plt.xlabel("Burası x ekseni")
plt.ylabel("Burası y ekseni")
plt.title("Burası başlık")
plt.legend("Burası Legend",loc='lower right')
plt.show()
# diğer legend lokasyonları: https://matplotlib.org/api/_as_gen/matplotlib.pyplot.legend.html
```

# Eksen Ölçek Araklıklarını Belirleme

```python
plt.plot()
plt.xlabel("Burası x ekseni")
plt.ylabel("Burası y ekseni")
plt.title("Burası başlık")
plt.legend("Burası Legend",loc='lower right')
plt.xticks([1, 2, 3, 4, 5])
#plt.xticks([1, 2, 3, 4, 5], ["1M", "2M", "3M", "4M", "5M"])
plt.yticks([1, 2, 3, 4, 5])
plt.show()
```

# Kanvas Genişliği

```python
plt.figure(figsize=(12,5)) # İlk değer x ekseni, ikinci değer y ekseni

plt.plot()
plt.xlabel("Burası x ekseni")
plt.ylabel("Burası y ekseni")
plt.title("Burası başlık")
plt.legend("Burası Legend",loc='lower right')
# plt.xticks([1, 2, 3, 4, 5])
plt.xticks([1, 2, 3, 4, 5], ["1M", "2M", "3M", "4M", "5M"])
plt.yticks([1, 2, 3, 4, 5])
plt.show()
```

```python
years = [1983, 1984, 1985, 1986, 1987]
total_populations = [8939007, 8954518, 8960387, 8956741, 8943721]

plt.figure(figsize=(9,6)) # Kanvas ebadı
plt.plot(years, total_populations)
plt.title("Year vs Population")
plt.xlabel("Year")
plt.ylabel("Total Population")
plt.show()
```
